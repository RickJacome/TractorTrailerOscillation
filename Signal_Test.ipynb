```python
import numpy as np
import matplotlib.pyplot as plt
t = np.linspace(0,100,100)
y = np.sin(t*2*np.pi)

mean = 0; std_dev = 0.2  
noise = np.random.normal(mean, std_dev, size=len(y))
y_n = y + noise

plt.plot(t,y_n)
```

```python
# Nisssan Sentra Specs
G = 9.81; m = 1281;  
# Actual lenght is 4.6 m
h = 1.5; h_cg = 0.4*h
L = 2.7; t1 = 1.5; t2 = 1.5 
a = 0.4*L;  b= 0.6*L
```

```python
a_x = y_n
F_z1 = (G*L*m - G*a*m - a_x*h_cg*m)/L
F_z2 = (G*a*m + a_x*h_cg*m)/L
F_x1 = a_x*m/2
F_x2 = a_x*m/2
```

```python
plt.plot(t,F_z1, label = 'Font Load'); plt.plot(t,F_z2, label = 'Rear Load')
plt.xlabel('Time (sec)'); plt.ylabel('Load (N)')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid()
```
